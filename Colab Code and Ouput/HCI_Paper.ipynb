```python
# ============================================================
# Research Topic:
# Explainable AI-Based Phishing Website Detection System
# for Non-Technical Users
#
# Dataset:
# Web Page Phishing Detection Dataset
# File:
# dataset_B_05_2020.csv
#
# Google Drive Path:
# /content/drive/MyDrive/HCI_Project/dataset_B_05_2020.csv
# ============================================================


# ============================================================
# 0. Install Required Libraries
# ============================================================

!pip install -q xgboost shap joblib


# ============================================================
# 1. Import Required Libraries
# ============================================================

import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib

from google.colab import drive

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report,
    ConfusionMatrixDisplay,
    RocCurveDisplay,
    PrecisionRecallDisplay
)

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    GradientBoostingClassifier
)

from xgboost import XGBClassifier
import shap


# ============================================================
# 2. Mount Google Drive
# ============================================================

drive.mount("/content/drive")


# ============================================================
# 3. Project Configuration
# ============================================================

RANDOM_STATE = 42
TEST_SIZE = 0.20

PROJECT_FOLDER = "/content/drive/MyDrive/HCI_Project"
DATASET_PATH = "/content/drive/MyDrive/HCI_Project/dataset_B_05_2020.csv"

OUTPUT_FOLDER = os.path.join(PROJECT_FOLDER, "Web_Phishing_Outputs")
os.makedirs(OUTPUT_FOLDER, exist_ok=True)

print("Dataset Path:", DATASET_PATH)
print("Output Folder:", OUTPUT_FOLDER)


# ============================================================
# 4. Load Dataset
# ============================================================

if not os.path.exists(DATASET_PATH):
    raise FileNotFoundError(
        "Dataset file not found. Please confirm that dataset_B_05_2020.csv "
        "is uploaded inside /content/drive/MyDrive/HCI_Project"
    )

df = pd.read_csv(DATASET_PATH)

# Clean column names
df.columns = [str(col).strip() for col in df.columns]

print("\nDataset loaded successfully.")
print("Dataset shape:", df.shape)

display(df.head())


# ============================================================
# 5. Dataset Overview
# ============================================================

print("\nColumn Names:")
print(df.columns.tolist())

print("\nDataset Information:")
df.info()

print("\nMissing Values:")
missing_values = df.isnull().sum().sort_values(ascending=False)
display(missing_values[missing_values > 0])

print("\nDuplicate Rows:", df.duplicated().sum())

print("\nTarget Column Distribution:")
print(df["status"].value_counts())


# ============================================================
# 6. Save Dataset Summary for Research Paper
# ============================================================

summary_path = os.path.join(OUTPUT_FOLDER, "dataset_summary.csv")
df.describe(include="all").transpose().to_csv(summary_path)

print("\nDataset summary saved at:")
print(summary_path)


# ============================================================
# 7. Target Variable Preparation
# ============================================================

# Original target column:
# status = legitimate / phishing
#
# Converted target:
# 0 = Legitimate Website
# 1 = Phishing Website

target_column = "status"

df[target_column] = df[target_column].astype(str).str.lower().str.strip()

y = df[target_column].map({
    "legitimate": 0,
    "phishing": 1
})

if y.isnull().sum() > 0:
    raise ValueError("Target column has unknown labels. Please check status column values.")

print("\nConverted Target Meaning:")
print("0 = Legitimate Website")
print("1 = Phishing Website")

print("\nConverted Target Distribution:")
print(y.value_counts())


# ============================================================
# 8. Feature Preparation
# ============================================================

# Drop text columns and target column
# url is dropped because ML models need numeric input.
# status is the target column.

X = df.drop(columns=["url", "status"], errors="ignore")

# Keep numeric features only
X = X.select_dtypes(include=[np.number]).copy()

# Replace infinite values with NaN
X = X.replace([np.inf, -np.inf], np.nan)

# Fill missing values with median
X = X.fillna(X.median(numeric_only=True))

# Remove duplicate feature columns if any
X = X.loc[:, ~X.columns.duplicated()]

print("\nFinal Feature Shape:", X.shape)
print("Total Missing Values After Cleaning:", X.isnull().sum().sum())

feature_list_path = os.path.join(OUTPUT_FOLDER, "feature_list.csv")
pd.DataFrame({"Feature": X.columns}).to_csv(feature_list_path, index=False)


# ============================================================
# 9. Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y
)

print("\nTraining Data Shape:", X_train.shape)
print("Testing Data Shape:", X_test.shape)


# ============================================================
# 10. Define Machine Learning Models
# ============================================================

models = {
    "Logistic Regression": Pipeline([
        ("scaler", StandardScaler()),
        ("classifier", LogisticRegression(
            max_iter=3000,
            random_state=RANDOM_STATE,
            class_weight="balanced"
        ))
    ]),

    "Decision Tree": DecisionTreeClassifier(
        random_state=RANDOM_STATE,
        class_weight="balanced",
        max_depth=None
    ),

    "Random Forest": RandomForestClassifier(
        n_estimators=300,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        class_weight="balanced"
    ),

    "Extra Trees": ExtraTreesClassifier(
        n_estimators=300,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        class_weight="balanced"
    ),

    "Gradient Boosting": GradientBoostingClassifier(
        random_state=RANDOM_STATE
    ),

    "XGBoost": XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=6,
        subsample=0.90,
        colsample_bytree=0.90,
        random_state=RANDOM_STATE,
        eval_metric="logloss",
        n_jobs=-1
    )
}


# ============================================================
# 11. Train and Evaluate Models
# ============================================================

results = []
trained_models = {}

for model_name, model in models.items():
    print("\n====================================================")
    print("Training Model:", model_name)
    print("====================================================")

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    roc_auc = roc_auc_score(y_test, y_prob)

    results.append({
        "Model": model_name,
        "Accuracy": round(accuracy, 5),
        "Precision_Phishing": round(precision, 5),
        "Recall_Phishing": round(recall, 5),
        "F1_Score_Phishing": round(f1, 5),
        "ROC_AUC": round(roc_auc, 5)
    })

    trained_models[model_name] = model

    print(classification_report(
        y_test,
        y_pred,
        target_names=["Legitimate", "Phishing"],
        zero_division=0
    ))


# ============================================================
# 12. Model Performance Table
# ============================================================

results_df = pd.DataFrame(results)

results_df = results_df.sort_values(
    by=["F1_Score_Phishing", "ROC_AUC"],
    ascending=False
)

print("\nFinal Model Performance Comparison:")
display(results_df)

results_csv_path = os.path.join(OUTPUT_FOLDER, "model_performance_results.csv")
results_df.to_csv(results_csv_path, index=False)

print("\nModel performance table saved at:")
print(results_csv_path)


# ============================================================
# 13. Cross-Validation for Best Models
# ============================================================

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=RANDOM_STATE
)

cv_results = []

for model_name in results_df["Model"].head(3):
    model = models[model_name]

    scores = cross_val_score(
        model,
        X,
        y,
        cv=cv,
        scoring="f1",
        n_jobs=-1
    )

    cv_results.append({
        "Model": model_name,
        "Mean_CV_F1": round(scores.mean(), 5),
        "Std_CV_F1": round(scores.std(), 5)
    })

cv_results_df = pd.DataFrame(cv_results)

print("\nCross-Validation Results for Top 3 Models:")
display(cv_results_df)

cv_results_path = os.path.join(OUTPUT_FOLDER, "cross_validation_results.csv")
cv_results_df.to_csv(cv_results_path, index=False)


# ============================================================
# 14. Select Best Model
# ============================================================

best_model_name = results_df.iloc[0]["Model"]
best_model = trained_models[best_model_name]

print("\nBest Model Selected:", best_model_name)

best_model_path = os.path.join(OUTPUT_FOLDER, "best_phishing_detection_model.pkl")
joblib.dump(best_model, best_model_path)

print("Best model saved at:")
print(best_model_path)


# ============================================================
# 15. Model Comparison Graphs
# ============================================================

plt.figure(figsize=(10, 6))
plt.bar(results_df["Model"], results_df["Accuracy"])
plt.xlabel("Machine Learning Models")
plt.ylabel("Accuracy")
plt.title("Accuracy Comparison of Machine Learning Models")
plt.xticks(rotation=30)
plt.tight_layout()

accuracy_plot_path = os.path.join(OUTPUT_FOLDER, "accuracy_comparison.png")
plt.savefig(accuracy_plot_path, dpi=300, bbox_inches="tight")
plt.show()


plt.figure(figsize=(10, 6))
plt.bar(results_df["Model"], results_df["F1_Score_Phishing"])
plt.xlabel("Machine Learning Models")
plt.ylabel("F1-Score for Phishing Class")
plt.title("F1-Score Comparison for Phishing Website Detection")
plt.xticks(rotation=30)
plt.tight_layout()

f1_plot_path = os.path.join(OUTPUT_FOLDER, "f1_score_comparison.png")
plt.savefig(f1_plot_path, dpi=300, bbox_inches="tight")
plt.show()


plt.figure(figsize=(10, 6))
plt.bar(results_df["Model"], results_df["ROC_AUC"])
plt.xlabel("Machine Learning Models")
plt.ylabel("ROC-AUC")
plt.title("ROC-AUC Comparison of Machine Learning Models")
plt.xticks(rotation=30)
plt.tight_layout()

auc_plot_path = os.path.join(OUTPUT_FOLDER, "roc_auc_comparison.png")
plt.savefig(auc_plot_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# 16. Confusion Matrix for Best Model
# ============================================================

best_y_pred = best_model.predict(X_test)
best_y_prob = best_model.predict_proba(X_test)[:, 1]

cm = confusion_matrix(y_test, best_y_pred)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Legitimate", "Phishing"]
)

disp.plot()
plt.title("Confusion Matrix - " + best_model_name)

confusion_matrix_path = os.path.join(OUTPUT_FOLDER, "confusion_matrix_best_model.png")
plt.savefig(confusion_matrix_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# 17. ROC Curve and Precision-Recall Curve
# ============================================================

RocCurveDisplay.from_predictions(y_test, best_y_prob)
plt.title("ROC Curve - " + best_model_name)

roc_curve_path = os.path.join(OUTPUT_FOLDER, "roc_curve_best_model.png")
plt.savefig(roc_curve_path, dpi=300, bbox_inches="tight")
plt.show()


PrecisionRecallDisplay.from_predictions(y_test, best_y_prob)
plt.title("Precision-Recall Curve - " + best_model_name)

pr_curve_path = os.path.join(OUTPUT_FOLDER, "precision_recall_curve_best_model.png")
plt.savefig(pr_curve_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# 18. Feature Importance
# ============================================================

def get_inner_model(model):
    if isinstance(model, Pipeline):
        return model.named_steps["classifier"]
    return model

inner_best_model = get_inner_model(best_model)

if hasattr(inner_best_model, "feature_importances_"):
    importance_values = inner_best_model.feature_importances_

elif hasattr(inner_best_model, "coef_"):
    importance_values = np.abs(inner_best_model.coef_[0])

else:
    importance_values = None


if importance_values is not None:
    feature_importance_df = pd.DataFrame({
        "Feature": X.columns,
        "Importance": importance_values
    }).sort_values(by="Importance", ascending=False)

    print("\nTop 20 Important Features:")
    display(feature_importance_df.head(20))

    feature_importance_path = os.path.join(OUTPUT_FOLDER, "feature_importance_results.csv")
    feature_importance_df.to_csv(feature_importance_path, index=False)

    plt.figure(figsize=(10, 8))
    plt.barh(
        feature_importance_df["Feature"].head(20),
        feature_importance_df["Importance"].head(20)
    )
    plt.xlabel("Importance")
    plt.ylabel("Feature")
    plt.title("Top 20 Important Features - " + best_model_name)
    plt.gca().invert_yaxis()
    plt.tight_layout()

    feature_importance_plot_path = os.path.join(
        OUTPUT_FOLDER,
        "top_20_feature_importance.png"
    )

    plt.savefig(feature_importance_plot_path, dpi=300, bbox_inches="tight")
    plt.show()

else:
    print("Feature importance is not available for the selected model.")


# ============================================================
# 19. SHAP Explainable AI Analysis
# ============================================================

print("\nRunning SHAP Explainable AI analysis...")

# Use a sample to make SHAP faster
X_shap = X_test.sample(n=min(700, len(X_test)), random_state=RANDOM_STATE)

try:
    if isinstance(best_model, Pipeline):
        classifier = best_model.named_steps["classifier"]
        scaler = best_model.named_steps["scaler"]
        X_shap_transformed = scaler.transform(X_shap)

        X_shap_for_plot = pd.DataFrame(
            X_shap_transformed,
            columns=X.columns
        )

        if hasattr(classifier, "coef_"):
            explainer = shap.LinearExplainer(classifier, X_shap_for_plot)
            shap_values = explainer.shap_values(X_shap_for_plot)
        else:
            raise ValueError("Pipeline model is not suitable for selected SHAP method.")

    else:
        classifier = best_model
        X_shap_for_plot = X_shap

        if hasattr(classifier, "feature_importances_"):
            explainer = shap.TreeExplainer(classifier)
            shap_values = explainer.shap_values(X_shap_for_plot)
        else:
            raise ValueError("Selected model is not compatible with TreeExplainer.")

    if isinstance(shap_values, list):
        shap_values_plot = shap_values[1]
    elif len(np.array(shap_values).shape) == 3:
        shap_values_plot = shap_values[:, :, 1]
    else:
        shap_values_plot = shap_values

    shap.summary_plot(
        shap_values_plot,
        X_shap_for_plot,
        show=False
    )

    plt.title("SHAP Summary Plot - " + best_model_name)
    plt.tight_layout()

    shap_summary_path = os.path.join(OUTPUT_FOLDER, "shap_summary_plot.png")
    plt.savefig(shap_summary_path, dpi=300, bbox_inches="tight")
    plt.show()

    print("SHAP summary plot saved at:")
    print(shap_summary_path)

except Exception as error:
    print("SHAP analysis could not be completed for the best model.")
    print("Reason:", error)


# ============================================================
# 20. Single Website Prediction Example
# ============================================================

sample = X_test.iloc[[0]]
sample_prediction = best_model.predict(sample)[0]
sample_probability = best_model.predict_proba(sample)[0]

print("\nSingle Website Prediction Example")

if sample_prediction == 1:
    print("Prediction: Phishing Website")
else:
    print("Prediction: Legitimate Website")

print("Phishing Probability:", round(sample_probability[1], 4))
print("Legitimate Probability:", round(sample_probability[0], 4))


# ============================================================
# 21. Human-Friendly Explanation for Non-Technical Users
# ============================================================

def human_friendly_explanation(row):
    explanations = []

    if "length_url" in row.index and row["length_url"] > 75:
        explanations.append("The website URL is unusually long.")

    if "ip" in row.index and row["ip"] == 1:
        explanations.append("The website uses an IP address instead of a normal domain name.")

    if "nb_dots" in row.index and row["nb_dots"] > 4:
        explanations.append("The URL contains many dots, which may indicate suspicious subdomains.")

    if "nb_hyphens" in row.index and row["nb_hyphens"] > 3:
        explanations.append("The URL contains many hyphens, which attackers often use to imitate real websites.")

    if "nb_at" in row.index and row["nb_at"] > 0:
        explanations.append("The URL contains the '@' symbol, which can hide the real destination.")

    if "https_token" in row.index and row["https_token"] == 1:
        explanations.append("The URL contains a suspicious HTTPS token.")

    if "domain_age" in row.index and row["domain_age"] < 6:
        explanations.append("The domain appears very new, which can be risky.")

    if "web_traffic" in row.index and row["web_traffic"] == 0:
        explanations.append("The website has very low or unknown web traffic.")

    if "google_index" in row.index and row["google_index"] == 0:
        explanations.append("The website does not appear clearly indexed by Google.")

    if "page_rank" in row.index and row["page_rank"] == 0:
        explanations.append("The website has a weak reputation score.")

    if len(explanations) == 0:
        explanations.append("No major suspicious rule-based feature was found, but the ML model still uses all features for prediction.")

    return explanations


sample_row = sample.iloc[0]
explanations = human_friendly_explanation(sample_row)

print("\nSimple Explanation for Non-Technical User:")
for explanation in explanations:
    print("-", explanation)


# ============================================================
# 22. Save Research Experiment Summary
# ============================================================

summary_text = f"""
Research Topic:
Explainable AI-Based Phishing Website Detection System for Non-Technical Users

Dataset:
dataset_B_05_2020.csv

Dataset Shape:
{df.shape}

Final Feature Shape:
{X.shape}

Target Meaning:
0 = Legitimate Website
1 = Phishing Website

Best Model:
{best_model_name}

Best Model Results:
{results_df.iloc[0].to_dict()}

Output Folder:
{OUTPUT_FOLDER}

Generated Files:
1. dataset_summary.csv
2. feature_list.csv
3. model_performance_results.csv
4. cross_validation_results.csv
5. accuracy_comparison.png
6. f1_score_comparison.png
7. roc_auc_comparison.png
8. confusion_matrix_best_model.png
9. roc_curve_best_model.png
10. precision_recall_curve_best_model.png
11. feature_importance_results.csv
12. top_20_feature_importance.png
13. shap_summary_plot.png
14. best_phishing_detection_model.pkl
"""

summary_file_path = os.path.join(OUTPUT_FOLDER, "research_experiment_summary.txt")

with open(summary_file_path, "w") as file:
    file.write(summary_text)

print("\nResearch experiment summary saved at:")
print(summary_file_path)


# ============================================================
# 23. Final Message
# ============================================================

print("\nAll outputs saved successfully in:")
print(OUTPUT_FOLDER)
```
